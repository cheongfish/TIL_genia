# 클래스(Class)
* 데이터와 기능(함수)을 함께 묶는 기능
* 함수와 같이 정의해서 사용함
* 함수를 동시에 여러번 사용할 때, 전역변수 편집을 할 경우 생기는 문제 해결 가능
    * 전역변수를 사용할 때: 전역변수를 바꿔가며 사용할 수 없다.
    * 클래스를 사용할 때: 함수 사용별로 변수를 초기화할 수 있다.

```python
# 예시 1
class Myclass:
    i = 12345  # <<  클래스에서 사용될 매개변수
    
    def f(self):
        return 'hello world'
Myclass.f(1)
```

```python
# 용례 2
class Complex:
    def __init__(self, realpart, imagpart): # 초기화 하면 변수를 저장할 수 있음
        self.r = realpart # self = 변수에 담아서 저장할 때 할당되는 변수
        self.i = imagpart
        # 다른 함수 선언 가능, 선언할 경우 r,i 사용가능함
        # 그리고, 아래 셀의 코드처럼 counter변수를 만들 경우, counter도 사용가능
        #complex 바깥에서 사용불가
x = Complex(3.0, -4.5)
x.r, x.i
```

```python
x.counter = 1
while x.counter < 10:
    x.counter = x.counter * 2
print(x.counter)
del x.counter # << 번수를 지움

```

# 데이터 전처리
* 데이터는 기본적으로 누락, 오차, 형태에 맞지 않는 값 등의 문제가 있다
* 데이터 전처리를 통해 해당요소 제거 가능
---
- 데이터 전처리의 종류
    * 데이터 클리닝
    * 데이터 통합
    * 데이터 변환
    * 데이터 축소
    * 데이터 이산화
    * 결측치 및 이상치 처리
    * 일관성 있는 데이터로 변환(표준화, 정규화)



## 데이터 클리닝
* 중복데이터 삭제
* 관련없는 데이터 삭제
* 구조적 문제 수정
    * 대소문자 일치
    * NA 표기 통일
* 이상치(outlier)처리 >> 처리하는 범위가 각각 다르므로 자신의 주장에 맞게 처리해야함

* 누락(NA)데이터 처리
    * 결측치 제거
    * 결측치를 특정한 값으로 대체
* 수정된 데이터 평가
    * 원본데이터와 수정된 데이터간 얼마나 왜곡 되었는지 측정


## 결측치, 이상치 처리
* 분석 불가능하거나 크게 영향을 주는 데이터 처리
* 결측치 처리
    * 제로필(0-fill)
        *  산술연산 가능, 데이터 처리할 때 간편함 , 다른값에 영향을 끼칠 수 있음, 데이터의 값이 클때 쓰면 상당히 왜곡됨
    * 평균값 
        *  통계분석시 결과에 크게 영향끼치지 않음, 분포가 밀집되게 함, 편차가 중요한 경우 평균값 기입시 문제 발생
    * 중간값 
        * 평균값의 장단점과 거의 비슷
    * 기계학습을 통한 대체 
        * 정밀한 데이터, 값을 대체하는 거 자체가 문제일 경우, 학습이 안된 분야인 경우 문제 발생함
        * 사용할 경우 원본데이터와 수정데이터간 벤치마킹 후 사용하는 것을 권장함
* 이상치 처리
    * 제거 
        *  통계분석시 유의미한 결과 도출, 데이터의 갯수가 줄기 때문에 불리함, 만약 의미있는 데이터가 제거된 경우 조작의 가능성
        * 신중하게 해야함
    * 대체
        * 분포가 유의미하고 샘플의 갯수가 적을때 효과적, 데이터가 왜곡되고 조작의 위험성 존재함
    * 기계학습을 통한 대체
        * 조작의 논란이 있을 수 있으므로, 신중하게 사용해야함
