# 데이터와 정보의 차이
- 데이터(data)
    * 단순히 모아놓은 자료의 총칭
    * 별도의 의미가 없는 기록에 그침
    * 데이터는 분석이 불가능하다
    * 기온, 인구수, 방문기록 등
- 정보(information)
    * 데이터를 가공해 특별한 목적을 띤다
    * 분석이 가능한 것이 정보
    * 연 평균기온, 특정 지역의 전입/전출 기록, 랜드마크별 방문기록

# 데이터 전처리
* 데이터는 기본적으로 누락, 오차, 형태에 맞지 않는 값 등의 문제가 있다
* 데이터 전처리를 통해 해당요소 제거 가능
* 절차적으로 하지 않고, 자의적으로 판단해서 진행함
* 버리거나 조정하거나 둘중에 하나로 처리함
* 클리닝을 제일 우선적으로 진행해야함, 간결하게 진행
---
- 데이터 전처리의 종류
    * 데이터 클리닝
    * 데이터 통합
    * 데이터 변환
    * 데이터 축소
    * 데이터 이산화
    * 결측치 및 이상치 처리
    * 일관성 있는 데이터로 변환(표준화, 정규화)



## 데이터 클리닝
* 중복데이터 삭제
* 관련없는 데이터 삭제
* 구조적 문제 수정 - 특정한 데이터 형태가 있을때, 내가 다루고자 하는 데이터의 형태에 맞지 않는 데이터
    * 대소문자 일치
    * NA 표기 통일
* 이상치(outlier)처리 >> 처리하는 범위가 각각 다르므로 자신의 주장에 맞게 처리해야함
    * 클리닝:이상치 삭제
    * 스케일링: 이상치 조정(평균값에 가깝게)
* 누락(NA)데이터 처리
    * 결측치 확인
    * 결측치 제거
    * 결측치를 특정한 값으로 대체
* 수정된 데이터 평가
    * 원본데이터와 수정된 데이터간 얼마나 왜곡 되었는지 측정
    * 전처리의 평가지표

## 데이터 통합
출처가 다양한 데이터를 단일화하는 과정

* 동일한 데이터를 통합
    - 두 데이터간 차이가 있는지 확인(형태의 차이)
* 같은 의미를 가지는 데이터 검토
* 같은 데이터를 포함한 정보처리
* 상위정보를 하위정보 형식으로 바꾸는 게 좋음

## 데이터 변환
 분석이 불가하거나 어려운 형태를 가능한 형태로 변환할 수 있다
* 단위변환
* (문자열의 경우) 파싱(.split())
* 형태 변환
* 여러 의미를 가진 데이터의 경우 분리 저장
    * 컬럼 하나에 한 종류의 데이터만 저장 되어 있는게 바람직함

## 데이터 축소
분석이 불가한 데이터를 저차원, 저용량으로 축소
* 차원 축소
    * 컬럼 개수를 줄이는 것
    * 데이터의 특징에 따라 다름
* 유효숫자 변경
    * 숫자 단위를 변경하는 것 (예시 7억원에서 원단위를 억원으로 바꾸는 것)
    * 득보다 실이 많을 때 안하는걸 권장
* 데이터 용량 축소
    * 서버에서 작업할 때 중요함
    * 메모리에서 오버플로우가 발생할 때
    * 작업속도가 중요한 경우
* 수량 감소 - 정보를 날리는 작업
    * 패러미터만 남기고 삭제 (값만 남기는 작업), 아는 내용은 중복에서 적지 않는 관점과 동일
    * 선형 회귀
    * 히스토그램
    * 클러스터링 
        * 데이터를 집단별로 묶는 방법, 개별값은 전부 버림
    * 샘플링
        * 표본조사와 비슷, 표본으로 추출한 데이터만 남기고, 나머지는 전부 버림
        * 신뢰도는 떨어지지만 적은 데이터로 대표성을 나타낼 수 있는 유의미한 결과를 도출할 수 있다는 점에서 경제적임

## 데이터 이산화 
연속형 데이터를 범주형 데이터로 변경 (바이너리화) 두개로 나누는것 <br>
(0,1),(10,20),(True,False) <br>
이산화가 인코딩 안에 들어가 있는 개념 <br>
* 기계학습 등에서 주요하게 사용
* 숫자가 크게 높거나 비슷한 값의 연관성 제거
* 컴퓨터가 판단할 수 없다는 특성에 근거

## 결측치, 이상치 처리
분석 불가능하거나 크게 영향을 주는 데이터 처리
* 결측치 처리
    * 제로필(0-fill)
        *  산술연산 가능, 데이터 처리할 때 간편함 , 다른값에 영향을 끼칠 수 있음, 데이터의 값이 클때 쓰면 상당히 왜곡됨
    * 평균값 
        *  통계분석시 결과에 크게 영향끼치지 않음, 분포가 밀집되게 함, 편차가 중요한 경우 평균값 기입시 문제 발생
    * 중간값 
        * 평균값의 장단점과 거의 비슷
    * 기계학습을 통한 대체 
        * 정밀한 데이터, 값을 대체하는 거 자체가 문제일 경우, 학습이 안된 분야인 경우 문제 발생함
        * 사용할 경우 원본데이터와 수정데이터간 벤치마킹 후 사용하는 것을 권장함
* 이상치 처리
    * 제거 
        * 통계분석시 유의미한 결과 도출, 데이터의 갯수가 줄기 때문에 불리함, 만약 의미있는 데이터가 제거된 경우 조작의 가능성
        * 신중하게 해야함
    * 대체
        * 분포가 유의미하고 샘플의 갯수가 적을때 효과적, 데이터가 왜곡되고 조작의 위험성 존재함
    * 기계학습을 통한 대체
        * 조작의 논란이 있을 수 있으므로, 신중하게 사용해야함

## 데이터 표준화, 정규화
데이터의 분포에 따른 특성 제거 또는 통계를 위한 변환
* 표준화(정규 분포 형태로 변화, Normalizing)
    * 데이터베이스의 정규화 개념과 구분
    * 데이터 범위 차이가 달라짐
    * 데이터 범위의 제한 없음
    * Z-Score
* 정규화(0-1 범위로 변화 등,Scaling)
    * 데이터의 범위 차이가 유지됨
    * 데이터 범위의 제한이 생김
    * min-max scaling등 (범위변환)

    * 데이터 범위의 제한이 생김
    * min-max scaling등 (범위변환)

```python
# 표준화
Age_std = round(df['Age_fill_0'].std(),2)
list2 = []
for i in df.index:
    a = (df['Age_fill_0'][i]/-df['Age_fill_0'].mean()) / df['Age_fill_0'].std()
    list2.append(round(a,2))
df['Age_fill_0_N'] = list2
df
```

```python
## 표준화 답안
# 정규분포가 아니라 정규분포를 따르는 것임(std만 확인하면 됨)
# sklearn 패키지에서 지원함
age_norm = (df['Age']- df['Age'].mean())/df['Age'].std()
age_norm.describe()
```

```python
# 답안
age_scaling = (df['Age']-df['Age'].min()) / (df['Age'].max() - df['Age'].min())
age_scaling.describe()
```

```python
# 정규화
Age_max =df['Age_fill_0'].max()
Age_min =df['Age_fill_0'].min()
list1 = []
for i in df.index:
    a = (df.loc[i,'Age_fill_0'] - Age_min)/(Age_max - Age_min)
    list1.append(round(a,2))
    
list1
df['Age_fill_0_minmax'] = list1   
```
