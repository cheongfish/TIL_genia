```python
## 답안 2
for i in df.index:
    if df.loc[i,'강좌정원수']:
        if df.loc[i,'강좌정원수']:
            print(df.loc[i,'강좌명'])
```

```python
for i in df.index: # 데이터프레임의 인덱스를 순회환다
    # all()함수 : 패러미터로 입력된 리스트에서, 구성요소 모두가 True인 경우 True , 아니면 False리턴
    #
    if all([df.loc[i,'강좌정원수']>=20, df.loc[i,'강좌정원수'] <=30]):
        print(df.loc[i,'강좌명'])
```

```python
df['강사명'].unique()
```

```python
for i in df.index: # df가 index를 순회함
    # any() 함수: 패러미터 모두 True 또는 하나 이상이 True일 경우 True를 리턴, 모두 False면 False리턴
    # >> 둘중에 하나가 True라면 True리턴
    # A in B : B 안에 A가 들어있을 경우 True, 없을 경우 False리턴
    if any([',' in df.loc[i,'강사명'], '외' in df.loc[i,'강사명']]):
        if '아동' in df.loc[i,'교육대상구분']:
            print(df.loc[i,'강좌명'])
    else:
        pass
```
