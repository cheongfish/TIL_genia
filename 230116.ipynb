# seaborn, matplotlib를 이용한 데이터 시각화

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid") # seaborn의 테마중 하나인 whitegrid로 나옴
# sns.set() seborn의 기본테마인 darkgrid로 나옴
# grid = 격자
```

```python
penguins = sns.load_dataset("penguins") # 컬럼별로 한가지 정보만 있는 데이터는 정형데이터

```

```python
g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
# catplot 카테고리 별로 확인할 수 있는 그래프
# hue =  x축의 값들을 구분할 수 있음
# plaette 그래프의 색깔 dark 기본값
# aplha는 불투명도
# 정형데이터를 인자로 넣어야함
```

```python
g.despine(left=True) # 축정보의 옵션, 안적어도 무방함
```

```python
g.set_axis_labels("", "Body mass (g)") # 축 라벨로 정함(표기)
```

```python
g.legend.set_title("") # 그래프의 범주표기
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

penguins = sns.load_dataset("penguins")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    errorbar="sd", palette="dark", alpha=1, height=6
)
g.despine()
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")

plt.show() # 위 코드들로만 작성후 실행시 그래프가 안나타날때 이 함수를 사용하면 됨
# 수량, 대소비교가 가능한 값들이 들어와야 함 = 양적변수여야함, barplot는 필연적으로 양적비교를 하기 때문에
# 카테고리별로 분류되어야함
# 그룹분류가 가능한 양적정보를 사용해서 이 그래프를 그려야함
# 그룹분류는 여러단계 이상으로 할 수 있음 (두단계까지 하는 걸 권장)
```

## 실습 2-1
* 위 그래프를 다음과 같이 수정해보자
1. x축의 이름을 식이요법을 하고, no fat = 무지방식단, low fat = 저지방 식단으로 바꿔보자
1. y축의 이름을 없앤다.
1. 그래프의 제목을 '식이요법에 따른 심박수 변화'로 하는데, 심박수 변화는 줄바꿈을 한 뒤 적는다
1. 해당 그래프를 다운로드 폴더에 저장한다 (이름: fat.jpg, 해상도 = 80dpi)

```python
list1 = []
for i in exercise.index:
    if exercise.loc[i,'diet'] == 'no fat':
        list1.append('무지방 식단')
    else:
        list1.append('저지방 식단')
exercise['diet']=list1
```

```python
# exercise['diet'] = exercise['diet'].apply(lambda x '무지방 식단' if x == 'no fat' else '저지방 식단')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4)) # 논문에서 (2,3), dpi  = 600으로 넣음
plt.rcParams['font.family']='Malgun Gothic'
sns.boxplot(data = exercise, x = 'diet', y = 'pulse')
plt.title('식이요법에 따른 \n심박수 변화')
plt.xlabel('식이요법')
plt.ylabel('')
plt.xticks([0,1], labels=['A','B']) # x축 바꾸기
# plt.yscale('log')
plt.ylim(100,160) # yscale조절, barplot에서 사용
# plt.savefig("C:/Users/user/Downloads/fat.jpg",dpi = 80)
```

## 실습 2-2
* diet분류에 따른 pulse 값의 차이가 있는지 알아보자
* 아래코드를 참고해서 boxplot과 t-test결과를 함께 출력한다

```python
from scipy import stats
exercise = sns.load_dataset('exercise')
m = exercise['diet'] =='low fat'
lowfat = exercise['pulse'][m]
nofat = exercise['pulse'][~m]
stats.ttest_ind(lowfat,nofat)
```

```python
pval=stats.ttest_ind(lowfat,nofat).pvalue
pval
```

```python
sns.boxplot(data = exercise, x = 'diet', y = 'pulse')
plt.xticks([0,1], labels=['무지방 식단','저지방 식단'])
plt.text(0.81,145,'p = %.4f' %pval) # plt.text(x축, y축, 'text')
```

# correlation analysis (상관관계 분석)
* 귀무가설 : 데이터 x와 y는 서로 선형적인 관계가 없다.

* p < 0.05이면 관계가 없는데, 있다고 할 확률이 5% 미만이므로, 관계가 있는 것으로 추정
* r (correlation coeffcient)의 값에 따라 상관관계가 얼마나 강한지 알아볼 수 있다
    1. 0.1 <  r <0.3 : 약한 상관관계
    1. 0.3 <  r <0.7 : 보통 수준의 상관관계
    1. 0.7 <  r <0.9 : 강한 상관관계
    1. r = 1 관계 : x=y
    1. 0.1> r : 상관관계 없음
    1. r이 음수인 경우, 위의 관계성에 모두 -1을 곱한다

```python
taxis  =sns.load_dataset('taxis')
taxis
```

```python
sns.scatterplot (x= taxis['distance'], y=taxis['fare'])
```

```python
sns.regplot (x= taxis['distance'], y=taxis['fare'])
```

```python
# pearson correlation analysis
# 정규성을 만족하고, 연속형형 데이터인 경우 
stats.pearsonr(taxis['distance'],taxis['fare'])
```

```python
# spearman correlation analysis
# 정규성을 만족하지 않거나, 카테고리(범주형) 데이터인 경우
# 애매하면 spearmanr correlation 사용
# 단 발표해야하면 정규성 검증후, pearson 검정 사용
stats.spearmanr(taxis['distance'],taxis['fare'])
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
sns.set_theme(style="darkgrid")
sns.scatterplot(x= diamonds['x'], y=diamonds['y'])
plt.text(9,55,'r = %.4f \np =%.4f'%(result[0],result[1]))
```

```python
sns.regplot (x= diamonds['x'], y=diamonds['y'],marker = '+')
plt.text(9,55,'r = %.4f \np =%.4f'%(result[0],result[1])) # marker옵션을 통해 점들의 모양을 바꿀 수 있음
```
